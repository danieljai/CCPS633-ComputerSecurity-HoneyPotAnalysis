```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)

df_all = pd.read_csv("sorted-AllTraffic.csv",sep='\t', lineterminator='\n')
```

replacing destination_port from None to 0

The standard ping command does not use TCP or UDP. It uses ICMP. To be more precise ICMP type 8 (echo message) and type 0 (echo reply message) are used. ICMP has no ports!

### Finding values

```python
df = df_all
```

```python
#df[(df['destination_port'] == "None") & (df['protocol'] != "None")].loc[:,'destination_port']  = 0
# fixing none numbers
df.loc[(df['destination_port'] == "None") & (df['protocol'] != "None"),'destination_port'] = 0
df.loc[(df['destination_port'] == "None") & (df['protocol'] == "None"),'destination_port'] = -1
```

```python
# converting all values to integer since some entries where in string
df['destination_port'] = df['destination_port'].astype(int)
```

```python
# does a value_count, assigning to a variable
s_port = df['destination_port'].value_counts()
```

```python
# top 25 ports that are being attacked.
print("List of Ports")
print(s_port.head(25).index)
print("Total count of ports being attacked in descending order")
print(s_port.head(25))
```

```python
dfp = pd.read_csv("commonports.csv")
dfp.iloc[0,1] = "ICMP"
```

```python
# list of the port description associated to the top 25 port numbers
dfp[dfp['port'].isin(list(s_port.head(25).index))].values
```
